```python
import pandas as pd
```

```python
dataset = pd.read_csv("Data.csv",encoding ='ISO-8859-1')
```

```python
#divide the data into train and test
train = dataset[dataset['Date']<'20150101']
test  = dataset[dataset['Date']>'20141231']
```

```python
#performing feature engineering
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex =True,inplace = True)
```

```python
#renaming the columns for the ease of access

list1 = [i for i in range(25)]

new_index = [str(i) for i in list1]
```

```python
data.columns
```

```python
data.columns = new_index
```

```python
data.columns
```

```python
data.head()
```

```python
# converting all the sentences into lower case
for index in new_index:
    data[index]=data[index].str.lower()
```

```python
data.head()
```

```python
# now all sentences are converted into lowercase
```

```python
# now we want to apply the nlp algorithm to this data so we will bunch this data together
```

```python
#creating an empty list first
bunched_lines = []
for row in range(len(data.index)):
    #we will use .join func
    bunched_lines.append(" ".join(str(x) for x in data.iloc[row,0:25]))
    
```

```python
bunched_lines[0]
```

```python
bunched_lines[1]
```

```python
# for displaying all the bunched lines
for i in range(len(data.index)):
    print(bunched_lines[i])
    print()
    print()
```

```python
# importing the bag of words algorithm 

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
count_vectorizer = CountVectorizer(ngram_range = (2,2))
traindataset = count_vectorizer.fit_transform(bunched_lines)
```

```python
# implementing the random forset classifier
randomclassifier = RandomForestClassifier(n_estimators =200,criterion ='entropy')
```

```python
randomclassifier.fit(traindataset,train['Label'])
```

```python
#predicting for the test data

test_transform =[]
for row in range(len(test.index)):
    test_transform.append(" ".join(str(x) for x in test.iloc[row,2:27]))
```

```python
test_dataset = count_vectorizer.transform(test_transform)
```

```python
predictions = randomclassifier.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix = confusion_matrix(test['Label'],predictions)
print(matrix)
```

```python
score = accuracy_score(test['Label'],predictions)
```

```python
print(score)
```

```python
report = classification_report(test['Label'],predictions)
print(report)
```
